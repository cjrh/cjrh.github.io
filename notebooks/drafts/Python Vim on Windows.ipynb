# How to setup Python integration in Vim on Windows

- Link to Haroogan's builds

By default, when Vim is started it will search through the folders listed in the `%PATH%` environment variable for the Python DLLs against which it was built. Currently, these are `python27.dll` for Python 2.x and `python34.dll` for Python 3.x. If these are found, a test of `:echo has('python')` for py2 and `:echo has('python3')` will return `1`.

If both Python 2.7 and 3.4 have been installed using the official installers, the python support in Vim will continue to work fine. However, ***if* you are using a Python that has not been "installed"**, such as a portable Python, or a different installer that allows you to omit the step of creating registry entries, the Python support in Vim will fail with this message:
```
The Python's site module could not be loaded.
```
This message means that the detected Python DLL, e.g. `python27.dll`, could not initialize correctly because the standard library could not be found.

The search order for finding the standard library proceeds in the following way:
- `*prefix*/lib`: where `*prefix*` is the folder of the executable, *not* the DLL. In our case, the executable is `vim.exe` or `gvim.exe`, and therefore the prefix is the folder that these executables are located in.
- `*prefix*/pythonXY.zip`: where the zip file contains the contents of the `Lib/` folder in a normal Python installation (and `X` and `Y` refer to the python version numbers.)
- The registry entries that would be created by a normal Python installation.

If you do not have the registry entries, `pythonXY.dll` will fail during initialization because `site.py` will fail to be imported. Looking at the steps above, we can resolve the problem by either creating a lib folder, or creating a zip file.  The zip file is the better option because it has a version-specific name, which means we can support both Python 2.7 and Python 3.4.

## Steps

0. Copy `python27.dll` and `python34.dll` into the same folder as `gvim.exe`
0. Assuming Python 2.7 has been installed normally, zip the **contents** of `C:\Python27\Lib` into a new file `python27.zip` into the same folder as `gvim.exe`
0. Assuming Python 3.4 has been installed normally, zip the **contents** of `C:\Python27\Lib` into a new file `python27.zip` into the same folder as `gvim.exe`

Note that only the contents are to be included in the zip file. If you view the zip file in an application like 7zip or some archive viewer, you should see, for example, files like `site.py` located in the root of the zip file structure.

## Caveats

It turns out that embedding Python into Vim does **not add `''` to `sys.path`** like a normal Python does, as explained in [this answer](http://stackoverflow.com/questions/2917645/embedding-python-in-c-having-problems-importing-local-modules). This small detail will interfere with the ability of the jedi-vim plugin to provide code completion in the current directory. To solve this, the easy way out is to add this code to your `vimrc`:

```python
python << EOF
import sys
if '' not in sys.path:
    sys.path.insert(0, '.')  # Tried '' but this one worked.
EOF
```
See also this [extensive article](http://blog.dscpl.com.au/2014/09/python-module-search-path-and-modwsgi.html) on how Python paths are controlled.
