# Koyeb Cloud Platform

I happened to test out a new cloud platform today: https://www.koyeb.com/ 

This is not an advertisement and I am not paid to write this or get any kickbacks and so on. I'm mainly documenting this so that I can refer back to it later if necessary.

## Hello World

By signing up you get a free account. It seems you can't actually deploy any of the LLM starter kits with a free account, but I managed to find the "Rust Template" for a web server.




Followed by selecting the Rust server template.



After the deploy screen, you will eventually get to the dashboard for your service.



Note that it has 0.1 vCPU and 512 MB memory.

Let's do a load test:
