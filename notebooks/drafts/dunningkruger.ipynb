```python
import matplotlib.pyplot as plt
```

```python
from random import choice
```

```python
actual = [choice(range(100)) for i in range(1000)]
```

```python
perceived = [choice(range(100)) for i in range(1000)]
```

```python
plt.plot(actual, perceived)
```

```python
aquartiles = [[], [], [], []]
pquartiles = [[], [], [], []]
for a, p in zip(actual, perceived):
    if 0 <= a < 25:
        aquartiles[0].append(a)
        pquartiles[0].append(p)
    elif 25 <= a < 50:
        aquartiles[1].append(a)
        pquartiles[1].append(p)

    elif 50 <= a < 75:
        aquartiles[2].append(a)
        pquartiles[2].append(p)

    elif 75 <= 75 < 100:
        aquartiles[3].append(a)
        pquartiles[3].append(p)
```
